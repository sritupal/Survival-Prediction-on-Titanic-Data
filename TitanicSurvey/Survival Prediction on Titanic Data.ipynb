# Bar Chart for categorical features
- pclass
- sex
- sibsp(#of siblings and apouse)
- parch(#of parents and children)
- embarked

```python
def bar_chart(feature):
    survived=train[train['Survived']==1][feature].value_counts()
    dead=train[train['Survived']==0][feature].value_counts()
    df=pd.DataFrame([survived,dead])
    df.index=['Survived','Dead']
    df.plot(kind='bar',stacked=True,figsize=(8,6))
```

```python
bar_chart('Sex')
print("Survived: \n",train[train['Survived']==1]['Sex'].value_counts())
print("\nDead: \n",train[train['Survived']==0]['Sex'].value_counts())
```

The chart confirms **Women more likely Survived than Men**.

```python
bar_chart('Pclass')
print("Survived: \n",train[train['Survived']==1]['Pclass'].value_counts())
print("\nDead: \n",train[train['Survived']==0]['Pclass'].value_counts())
```

The first chart confirms 1st class more likely survived than other class<br>
The second chart confirms 3rd more likely dead than other class

```python
bar_chart('SibSp')
print("Survived: \n",train[train['Survived']==1]['SibSp'].value_counts())
print("\nDead: \n",train[train['Survived']==0]['SibSp'].value_counts())
```

The first chart confirms a **person aboarded with more than 2 siblings or spouse** more likely survived.<br>
The second chart confirms a **person aboarded without siblings or spouse** more likely dead.

```python
bar_chart('Parch')
print("Survived: \n",train[train['Survived']==1]['Parch'].value_counts())
print("\nDead: \n",train[train['Survived']==0]['Parch'].value_counts())
```

The first chart confirms a **person aboarded with more than 2 parents or children more likely survived**.<br>
The second chart confirms a **person aboarded alone more likely dead**.

```python
bar_chart('Embarked')
print("Survived: \n",train[train['Survived']==1]['Embarked'].value_counts())
print("\nDead: \n",train[train['Survived']==0]['Embarked'].value_counts())
```

The chart confirms a **person aborded from C** slightly more likely survived.<br>
The chart confirms a **person aborded from Q** more likely dead.<br>
The chart confirms a **person aborded from S** more likely dead.

### Cross Validation(k-fold)

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold=KFold(n_splits=10,shuffle=True,random_state=0)
```

```python
clf=KNeighborsClassifier(n_neighbors=13)
scoring='accuracy'
score=cross_val_score(clf,train_data,target,cv=k_fold,n_jobs=1,scoring=scoring)
print(score)
```

```python
#learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
clf = [KNeighborsClassifier(n_neighbors = 13),DecisionTreeClassifier(),
       RandomForestClassifier(n_estimators=13),GaussianNB(),SVC(),ExtraTreeClassifier(),
      GradientBoostingClassifier(n_estimators=10, learning_rate=1,max_features=3, max_depth =3, random_state = 10),AdaBoostClassifier(),ExtraTreesClassifier()]
def model_fit():
    scoring = 'accuracy'
    for i in range(len(clf)):
        score = cross_val_score(clf[i], train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
        print("Score of Model",i,":",round(np.mean(score)*100,2))
#     round(np.mean(score)*100,2)
#     print("Score of :\n",score)
model_fit()
```

```python
clf1=SVC()
clf1.fit(train_data,target)
test
test_data=test.drop(['Survived','PassengerId'],axis=1)
prediction=clf1.predict(test_data)
#test_data
```
